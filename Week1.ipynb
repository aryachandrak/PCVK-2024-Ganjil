1. Load Library

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

2. Menampilkan Gambar

```python
# membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://scontent-xsp1-2.xx.fbcdn.net/v/t39.30808-6/309552767_441312951429711_5018110099174612061_n.jpg?_nc_cat=110&ccb=1-7&_nc_sid=cc71e4&_nc_eui2=AeEZF6MS4fZLFcLyu0vKeQjkb63LksB2CaVvrcuSwHYJpboYVd8aKtGLe7Jk83zkaUYzgMgovvRMctszdElWIITG&_nc_ohc=KkPSe8-Thi4Q7kNvgFt3lw4&_nc_ht=scontent-xsp1-2.xx&oh=00_AYAg9kETTsy0g_o5AzFZ930cF9vuY5EKe-YlKe1FXAcZSg&oe=66E4CE79"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                            # read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5)   # resize image
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)    # convert color to RGB
  final_frame = cv.hconcat((image, image_2))        # concat image
  cv2_imshow(final_frame)                           # show image
  print("\n")


```

```python
panjang = image_2.shape[0]
lebar = image_2.shape[1]
print("Resolusi image panjang x lebar = ", panjang, " x ", lebar)
cv2_imshow(image_2)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(lebar):
  image_3[int((panjang)/2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Tugas Kelompok

```python
image_ktm1 = cv.imread("KTM_Ricky.jpg")
samping = image_ktm1.shape[0]
bawah = image_ktm1.shape[1]
print("Resolusi image sisi samping x sisi bawah = ", samping, " x ", bawah)
image_ktm1 = cv.resize(image_ktm1, (383, 640))
samping_edit = image_ktm1.shape[0]
bawah_edit = image_ktm1.shape[1]
print("Resolusi image sisi samping edit x sisi bawah edit = ", samping_edit, " x ", bawah_edit)
image_sensorKtm1 = image_ktm1.copy()

for y in range(400, 510):
  image_sensorKtm1[y, 335:360] = [0, 0, 0] #sensor nim

for y in range(15, 130):
  image_sensorKtm1[y, 180:300] = [0, 0, 0] #sensor bar code

final_frame = cv.hconcat((image_ktm1, image_sensorKtm1))
cv2_imshow(final_frame)
```

```python
image_ktm2 = cv.imread("KTM_Armand.jpg")
samping = image_ktm2.shape[0]
bawah = image_ktm2.shape[1]
print("Resolusi image sisi samping x sisi bawah = ", samping, " x ", bawah)
image_ktm2 = cv.resize(image_ktm2, (414, 640))
samping_edit = image_ktm2.shape[0]
bawah_edit = image_ktm2.shape[1]
print("Resolusi image sisi samping edit x sisi bawah edit = ", samping_edit, " x ", bawah_edit)
image_sensorKtm2 = image_ktm2.copy()

for y in range(400, 500):
  image_sensorKtm2[y, 365:390] = [0, 0, 0] #sensor nim

for y in range(5, 120):
  image_sensorKtm2[y, 200:320] = [0, 0, 0] #sensor bar code

final_frame = cv.hconcat((image_ktm2, image_sensorKtm2))
cv2_imshow(final_frame)
```
