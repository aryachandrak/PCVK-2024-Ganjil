```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

2. Menampilkan Gambar

```python
# membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://scontent-xsp1-2.xx.fbcdn.net/v/t39.30808-6/309552767_441312951429711_5018110099174612061_n.jpg?_nc_cat=110&ccb=1-7&_nc_sid=cc71e4&_nc_eui2=AeEZF6MS4fZLFcLyu0vKeQjkb63LksB2CaVvrcuSwHYJpboYVd8aKtGLe7Jk83zkaUYzgMgovvRMctszdElWIITG&_nc_ohc=KkPSe8-Thi4Q7kNvgFt3lw4&_nc_ht=scontent-xsp1-2.xx&oh=00_AYAg9kETTsy0g_o5AzFZ930cF9vuY5EKe-YlKe1FXAcZSg&oe=66E4CE79"]

# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                            # read image
  image = cv.resize(image, (500,350), fx=0.5, fy=0.5)   # resize image
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)    # convert color to RGB
  final_frame = cv.hconcat((image, image_2))              # concat image
  cv2_imshow(final_frame)                           # show image
  print("\n")


```
